```python
# Flickrë¡œ ì‚¬ì§„ ê²€ìƒ‰í•´ì„œ ë‹¤ìš´ë¡œë“œí•˜ê¸°
from flickrapi import FlickrAPI
from urllib.request import urlretrieve
from pprint import pprint
import os, time, sys

# AP í‚¤ ì§€ì •í•˜ê¸°--- ( â€» 1)
# key = "<ìì‹ ì˜ ê²ƒì„ ì…ë ¥í•´ì£¼ì„¸ìš”>"
# secret = "<ìì‹ ì˜ ê²ƒì„ ì…ë ¥í•´ì£¼ì„¸ìš”>"
key = "1eb22788a01f7bace9649c51c3de8c03"
secret = "87b0bdb8d6b17cc5"
wait_time = 1 # ëŒ€ê¸° ì‹œê°„(ì´ˆ)

# í•œ ë²ˆì— ëŒ€ëŸ‰ì˜ ì´ë¯¸ì§€ë¥¼ ë‚´ë ¤ ë°›ìœ¼ë©´ Flickr ì„œë²„ì— ë¶€í•˜ê°€ ê±¸ë¦´ ìˆ˜ ìˆê¸°ì— 1ì´ˆ ê°„ê²©ìœ¼ë¡œ ë‚´ë ¤ë°›ê³ ì
```

```python
# Flickr APIë¡œ ì‚¬ì§„ ê²€ìƒ‰í•˜ê¸° 
def go_download(keyword, dir):
    # ì €ì¥ ê²½ë¡œ ì§€ì •í•˜ê¸°
    savedir = "./image/" + dir
    if not os.path.exists(savedir):
        os.mkdir(savedir)
    # APIë¥¼ ì‚¬ìš©í•´ì„œ ë‹¤ìš´ë¡œë“œí•˜ê¸° 
    flickr = FlickrAPI(key, secret, format='parsed-json')
    res = flickr.photos.search(
      text = keyword,     # í‚¤ì›Œë“œ
      per_page = 300,     # ê²€ìƒ‰í•  ê°œìˆ˜ ( Flickrì˜ ê²€ìƒ‰ APIì—ì„œ ìµœëŒ€ 300ê°œ ì´ë¯¸ì§€ ì œê³µ )
      media = 'photos',   # ì‚¬ì§„ ê²€ìƒ‰
      sort = "relevance", # í‚¤ì›Œë“œ ê´€ë ¨ë„ ìˆœì„œ
      safe_search = 1,    # ì•ˆì „ ê²€ìƒ‰
      extras = 'url_q, license')
    # ê²°ê³¼ í™•ì¸í•˜ê¸°
    photos = res['photos']
    pprint(photos)
    try:
      # 1ì¥ì”© ë‹¤ìš´ë¡œë“œí•˜ê¸°
      for i, photo in enumerate(photos['photo']):
        url_q = photo['url_q']
        filepath = savedir + '/' + photo['id'] + '.jpg'
        if os.path.exists(filepath): continue
        print(str(i + 1) + ":download=", url_q)
        urlretrieve(url_q, filepath)
        time.sleep(wait_time)
    except:
      import traceback
      traceback.print_exc()

```
